```python
# Step 1: Import required libraries
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec

# Step 2: Download tokenizer data
nltk.download('punkt')
nltk.download('punkt_tab') # Added to download the missing resource

corpus = [
    "machine learning is evolving",
    "AI is changing"
]

tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

print("Tokenized Corpus:")
for i, tokens in enumerate(tokenized_corpus):
    print(f"Sentence {i+1}: {tokens}")

model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=2,
    min_count=1,
    workers=1,
    sg=0
)
print("\nWord Embeddings:\n")
for word in model.wv.index_to_key:
    print(f"{word}: {model.wv[word]}")
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output.last_hidden_state
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(
        input_mask_expanded.sum(1), min=1e-9
    )
def get_sentence_embedding(sentence):
    encoded_input = tokenizer(
        sentence,
        padding=True,
        truncation=True,
        return_tensors='pt'
    )
    with torch.no_grad():
        model_output = model(**encoded_input)
    sentence_embedding = mean_pooling(
        model_output,
        encoded_input['attention_mask']
    )
    return sentence_embedding.numpy()
def text_similarity(text1, text2):
    emb1 = get_sentence_embedding(text1)
    emb2 = get_sentence_embedding(text2)
    similarity = cosine_similarity(emb1, emb2)
    return similarity[0][0]
text_a = "Machine learning is transforming technology"
text_b = "AI is changing how machines learn"
score = text_similarity(text_a, text_b)
print("Similarity Score:", score)
```

```python
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity
model_name = "sentence-transformers/all-MiniLM-L6-v2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output.last_hidden_state
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(
        input_mask_expanded.sum(1), min=1e-9
    )
def get_sentence_embedding(sentence):
    encoded_input = tokenizer(
        sentence,
        padding=True,
        truncation=True,
        return_tensors='pt'
    )
    with torch.no_grad():
        model_output = model(**encoded_input)
    sentence_embedding = mean_pooling(
        model_output,
        encoded_input['attention_mask']
    )
    return sentence_embedding.numpy()
def text_similarity(text1, text2):
    emb1 = get_sentence_embedding(text1)
    emb2 = get_sentence_embedding(text2)
    similarity = cosine_similarity(emb1, emb2)
    return similarity[0][0]

# Get input from user
text_a = "The quick brown fox jumps over the lazy dog."
text_b = "A fast brown fox leaps over a lethargic canine."

score = text_similarity(text_a, text_b)
print("Similarity Score:", score)

if score > 0.5:
    print("The sentences are similar.")
else:
    print("The sentences are not very similar.")
```

```python
from transformers import pipeline
import numpy as np
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
texts = [
    "I love this sunny weather!",
    "This movie was utterly boring.",
    "The food was delicious and the service was excellent.",
    "I'm feeling really sad today.",
    "What a fantastic experience!"
]
results = sentiment_pipeline(texts)
for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.2f}")
    print("-" * 50)
```
